```python
#cleaning and wrangling housing data
import pandas as pd
import numpy as np 

#reading the csv files
housedata = pd.read_csv('housing_neat.csv')

#normalising the suburb names into standardised lowercase
cleanhouse = pd.DataFrame(housedata)
rate = housedata['growth_pa']
region_id = pd.Series(cleanhouse['locality'])

#loop for standardising data region into lowercase
for i in range(len(region_id)):
    region_id[i] = str.lower(region_id[i])

cleanhouse['locality'] = region_id



#function to detect nulls
def valid_value(value):
    if type(value) == float or type(value) == int:
        return True
    else:
        return False
    
    
#function to fill in all the missing values
def fill_blanks(cleanhouse):
    nullvalue = False
    
    #looping through the rows 
    for i in range(len(cleanhouse)):
        
        #checking if there exists a growth_pa rate for this row
        ############################################################
        #Note that the growth_pa method of calculations produced uneven and often biased results and 
        #was replaced with a slightly less noisy method of replacing null cells with averages for the row
        if (valid_value(cleanhouse.ix[i, 'growth_pa'])):
            pa = cleanhouse.ix[i, 'growth_pa']
        else:
            pa = 0
            
        #initialising some variables, number of years in data set and zeros if detected
        nyears = len(cleanhouse.ix[0])-2
        zerocount = 0
        #mean = (len(cleanhouse.ix[i])).mean
        
        #checking if there is a zero value, counting them as things go
        for j in range(nyears):
            if(cleanhouse.ix[i, j+1]) == 0:
                nullvalue = True
                zerocount+= 1
            
        #replacing the zero values by the mean of the present values in the row
        if nullvalue:
            #calculate mean
            mean = (sum(cleanhouse.ix[i][1:-1]))/zerocount
            #loop through the row again to replace all the zeros with mean values
            for k in range(nyears):
                if ((cleanhouse.ix[i, k+1]) == 0):
                    cleanhouse.loc[i,str(2005+k)]= int(mean)

                        
        nullvalue = False    
                
    return cleanhouse   


cleanhouse = fill_blanks(cleanhouse)
cleanhouse.to_csv('cleanhouse.csv')
cleanhouse
```
